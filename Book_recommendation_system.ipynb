# **Popularity Based Recommendation System**

A popularity-based recommendation system is a simple approach to providing recommendations based on the overall popularity or popularity trends of items. It doesn't consider individual user preferences but rather suggests items that are generally popular or trending among users.

```python
## display the books with the highest average ratings
ratings.merge(books,on='ISBN')
```

```python
ratings.merge(books,on='ISBN').shape
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python

ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
##What is inplace true?
##Image result for inplace= true
##When inplace = True , the data is modified in place, which means it will return nothing and the dataframe is now updated. When inplace = False , 
#which is the default, then the operation is performed and it returns a copy of the object. 
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()

```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df)
popular_df
```

```python
#filters the DataFrame to include only the rows where the 'num_ratings' column has a value greater than or equal to 250.
popular_df[popular_df['num_ratings']>=250]
```

```python
#filters the DataFrame popular_df to include only the rows where the 'num_ratings' column has a value greater than or equal to 250. 
#Then, it sorts the filtered DataFrame in descending order based on the 'avg_ratings' column.
popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
## we need Top 50 books
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
popular_df
```

```python
# merge operation between the DataFrames popular_df and books based on the common column 'Book-Title'. 
# Then, remove any duplicate rows based on the 'Book-Title' column, resulting in a DataFrame with unique 'Book-Title' entries.
popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
# merge  between the DataFrames popular_df and books based on the common column 'Book-Title'. 
#Then, remove any duplicate rows based on the 'Book-Title' column and selects specific columns from the resulting DataFrame.
popular_df= popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

# **Collaborative Filtering Based Recommender System**

A collaborative filtering-based recommender system is an approach that suggests items to users based on the preferences and behavior of similar users. It utilizes the idea that users with similar tastes or preferences in the past are likely to have similar preferences in the future.

```python
ratings_with_name
```

```python
# group the DataFrame ratings_with_name by the 'User-ID' column and counts the number of occurrences of 'Book-Rating' for each unique 'User-ID'.
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200

best_users = x[x].index
```

```python
#filter the DataFrame ratings_with_name to include only the rows where the 'User-ID' column value is present in the list best_users.
filtered_rating =ratings_with_name[ratings_with_name['User-ID'].isin(best_users)]
```

```python
#Group the DataFrame filtered_rating by the 'Book-Title' column and counts the number of occurrences of 'Book-Rating' for each unique book title. 
#Then, it applies a condition to check if the count is greater than or equal to 50. 
filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50

famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
cosine_similarity(pt)
```

```python
similarity_scores =  cosine_similarity(pt)
similarity_scores.shape
```

```python
np.where(pt.index=='4 Blondes')[0][0]
```

```python
sorted(list(enumerate(similarity_scores[40])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
## Recommend Function
def recommend(book_name):
    #index fetch
    # index = np.where(pt.index=='4 Blondes')[0][0]
        index = np.where(pt.index==book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

        for i in  similar_items:
            print(pt.index[i[0]])
```

```python
recommend('4 Blondes')
```

```python
recommend('2nd Chance')
```
